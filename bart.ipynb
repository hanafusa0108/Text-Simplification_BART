```python
# テキストファイルのパス
file_path = 'valid-ja-simp.pred'

# ファイルを読み込んで単語を1行ずつくっつけるための空の文字列を作成
no_spaces_text = ""
list = []

# ファイルを開いて1行ずつ読み込み、単語をくっつける
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        # 行末の改行文字を削除してから単語をくっつける
        no_spaces_text = line.strip().replace(" ", "")
        #joined_text += line.strip() + " "
        list.append(no_spaces_text)

# リストをテキストファイルに書き込む
with open(file_path, 'w', encoding='utf-8') as file:
    for item in list:
        file.write(item + '\n')
```

```python
# テキストファイルのパス
file_path = 'ja-simp.pred'

# ファイルを読み込んで単語を1行ずつくっつけるための空の文字列を作成
no_spaces_text = ""
list = []

# ファイルを開いて1行ずつ読み込み、単語をくっつける
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        # 行末の改行文字を削除してから単語をくっつける
        no_spaces_text = line.strip().replace(" ", "")
        #joined_text += line.strip() + " "
        list.append(no_spaces_text)

# リストをテキストファイルに書き込む
with open(file_path, 'w', encoding='utf-8') as file:
    for item in list:
        file.write(item + '\n')
```

```python
# MeCabとipadicのインポート
import MeCab
import ipadic

# 辞書の初期化 + 分かち書き
tagger = MeCab.Tagger(ipadic.MECAB_ARGS + " -Owakati")

# トークナイズ
def tokenize(fname):
    fin = open(fname, "r")
    sentences = [tagger.parse(line.strip()) for line in fin]
    fin.close()
    return sentences
sentences = tokenize("ja-simp.pred")

# ファイルに書き込み
fout = open("test-bart-base-model.txt", "w")
for x in sentences:
    fout.write(str(x) + "\n")
fout.close()
```
